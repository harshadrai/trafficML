# This code creates a bar graph representing the contributing factors for traffic accidents in New York City

Import packages

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

Import the file as a dataframe

```python
df=pd.read_csv("SUPER_CLEANED_NYPD_TRAFFIC_COLLISION.csv")
```

Create a dictionary called causes and store the keys as the different contributing factors and their corresponding values as the number of accidents caused due to this reason

```python
causes={}
for index, row in df.iterrows():
    if row['CONTRIBUTING FACTOR VEHICLE 1'] in causes:
        causes[str(row['CONTRIBUTING FACTOR VEHICLE 1'])] += 1
    else:
        causes[str(row['CONTRIBUTING FACTOR VEHICLE 1'])] = 1
```

Delete keys that correspond nan or Unspecified from the dictionary

```python
del causes['nan']
```

```python
del causes['Unspecified']
```

Set the size of the figure accordingly as there are a number of causes and the visualization would require a large figure

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 18
fig_size[1] = 18
plt.rcParams["figure.figsize"] = fig_size
```

The next cell plots the bar graph

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm, matplotlib.cm as cm, matplotlib.patches as mpatches

plt.bar(range(len(causes)), causes.values())
plt.xticks(range(len(causes)), causes.keys(),rotation=30, rotation_mode='anchor', ha='right')
plt.title("Bar_Graphs_for_Contributing_Factors.png")
plt.savefig("Bar_Graphs_for_Contributing_Factors.png", dpi=1200)
plt.show()
```
