# This code clusters and visualizes traffic fatalities for different Boroughs in New York City based on different years from 2012-2016

Import packages and the file as a dataframe

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime

df = pd.read_csv('SUPER_CLEANED_NYPD_TRAFFIC_COLLISION.csv')
```

```python
#The list on 0 index in each year is the Latitude
#The list on 1 index in each year is the Longitude

Boroughs={'MANHATTAN':{2012:[[],[]], 2013:[[],[]],2014:[[],[]], 2015:[[],[]], 2016:[[],[]], 2017:[[],[]]}, 'BROOKLYN':{2012:[[],[]], 2013:[[],[]],2014:[[],[]], 2015:[[],[]], 2016:[[],[]], 2017:[[],[]]}, 'QUEENS':{2012:[[],[]], 2013:[[],[]],2014:[[],[]], 2015:[[],[]], 2016:[[],[]], 2017:[[],[]]}, 'BRONX':{2012:[[],[]], 2013:[[],[]],2014:[[],[]], 2015:[[],[]], 2016:[[],[]], 2017:[[],[]]}, 'STATEN ISLAND':{2012:[[],[]], 2013:[[],[]],2014:[[],[]], 2015:[[],[]], 2016:[[],[]], 2017:[[],[]]}}

for index, row in df.iterrows():
    a=int(row['NUMBER OF PERSONS KILLED'])
    if ((a>0) and not(np.isnan(row['LATITUDE']) and np.isnan(row['LONGITUDE']))):
        if row['BOROUGH']=='MANHATTAN':
            if int(pd.to_datetime(row['DATE']).year)==2016:
                Boroughs['MANHATTAN'][2016][0].append(float(row['LATITUDE']))
                Boroughs['MANHATTAN'][2016][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2015:
                Boroughs['MANHATTAN'][2015][0].append(float(row['LATITUDE']))
                Boroughs['MANHATTAN'][2015][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2014:
                Boroughs['MANHATTAN'][2014][0].append(float(row['LATITUDE']))
                Boroughs['MANHATTAN'][2014][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2013:
                Boroughs['MANHATTAN'][2013][0].append(float(row['LATITUDE']))
                Boroughs['MANHATTAN'][2013][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2012:
                Boroughs['MANHATTAN'][2012][0].append(float(row['LATITUDE']))
                Boroughs['MANHATTAN'][2012][1].append(float(row['LONGITUDE']))
        elif row['BOROUGH']=='BROOKLYN':
            if int(pd.to_datetime(row['DATE']).year)==2016:
                Boroughs['BROOKLYN'][2016][0].append(float(row['LATITUDE']))
                Boroughs['BROOKLYN'][2016][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2015:
                Boroughs['BROOKLYN'][2015][0].append(float(row['LATITUDE']))
                Boroughs['BROOKLYN'][2015][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2014:
                Boroughs['BROOKLYN'][2014][0].append(float(row['LATITUDE']))
                Boroughs['BROOKLYN'][2014][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2013:
                Boroughs['BROOKLYN'][2013][0].append(float(row['LATITUDE']))
                Boroughs['BROOKLYN'][2013][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2012:
                Boroughs['BROOKLYN'][2012][0].append(float(row['LATITUDE']))
                Boroughs['BROOKLYN'][2012][1].append(float(row['LONGITUDE']))
        elif (row['BOROUGH']=='QUEENS'):
            if int(pd.to_datetime(row['DATE']).year)==2016:
                Boroughs['QUEENS'][2016][0].append(float(row['LATITUDE']))
                Boroughs['QUEENS'][2016][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2015:
                Boroughs['QUEENS'][2015][0].append(float(row['LATITUDE']))
                Boroughs['QUEENS'][2015][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2014:
                Boroughs['QUEENS'][2014][0].append(float(row['LATITUDE']))
                Boroughs['QUEENS'][2014][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2013:
                Boroughs['QUEENS'][2013][0].append(float(row['LATITUDE']))
                Boroughs['QUEENS'][2013][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2012:
                Boroughs['QUEENS'][2012][0].append(float(row['LATITUDE']))
                Boroughs['QUEENS'][2012][1].append(float(row['LONGITUDE']))
        elif (row['BOROUGH']=='BRONX'):
            if int(pd.to_datetime(row['DATE']).year)==2016:
                Boroughs['BRONX'][2016][0].append(float(row['LATITUDE']))
                Boroughs['BRONX'][2016][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2015:
                Boroughs['BRONX'][2015][0].append(float(row['LATITUDE']))
                Boroughs['BRONX'][2015][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2014:
                Boroughs['BRONX'][2014][0].append(float(row['LATITUDE']))
                Boroughs['BRONX'][2014][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2013:
                Boroughs['BRONX'][2013][0].append(float(row['LATITUDE']))
                Boroughs['BRONX'][2013][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2012:
                Boroughs['BRONX'][2012][0].append(float(row['LATITUDE']))
                Boroughs['BRONX'][2012][1].append(float(row['LONGITUDE']))
        elif (row['BOROUGH']=='STATEN ISLAND'):
            if int(pd.to_datetime(row['DATE']).year)==2016:
                Boroughs['STATEN ISLAND'][2016][0].append(float(row['LATITUDE']))
                Boroughs['STATEN ISLAND'][2016][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2015:
                Boroughs['STATEN ISLAND'][2015][0].append(float(row['LATITUDE']))
                Boroughs['STATEN ISLAND'][2015][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2014:
                Boroughs['STATEN ISLAND'][2014][0].append(float(row['LATITUDE']))
                Boroughs['STATEN ISLAND'][2014][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2013:
                Boroughs['STATEN ISLAND'][2013][0].append(float(row['LATITUDE']))
                Boroughs['STATEN ISLAND'][2013][1].append(float(row['LONGITUDE']))
            elif int(pd.to_datetime(row['DATE']).year)==2012:
                Boroughs['STATEN ISLAND'][2012][0].append(float(row['LATITUDE']))
                Boroughs['STATEN ISLAND'][2012][1].append(float(row['LONGITUDE']))
```

We cluster the fatalities for different Boroughs and years and visualize them in the next cell using the DBSCAN algorithm. The DBSCAN library is imported from the sklearn.cluster package. We specify the minimum radius to be 1.2kms and the minimum number of fatalities to be 2. This is user dependent and can be changed accordingly when the function is called.

```python
def clustering_borough_fatality(kms, min_no):
    import matplotlib.pyplot as plt
    import pandas as pd
    import numpy as np
    import datetime
    borough_name=input("What Borough do you want to cluster (MANHATTAN, BROOKLYN, QUEENS, BRONX, STATEN ISLAND: ")
    if borough_name in Boroughs:
        year=input("What year would you like to cluster: (2012, 2013, 2014, 2015, 2016)")
        year=int(year)
        if year in Boroughs[borough_name]:
            final_list=[]
            for j in range(len(Boroughs[borough_name][year][0])):
                my_list=[]
                for i in range(len(Boroughs[borough_name][year])):
                    my_list.append(Boroughs[borough_name][year][i][j])
                final_list.append(my_list)


            coords = np.array(np.radians(np.array(final_list)))

            #The DBSCAN clustering part starts from here


            import pandas as pd, numpy as np, matplotlib.pyplot as plt, time
            from sklearn.cluster import DBSCAN
            from sklearn import metrics
            from geopy.distance import great_circle


            kms_per_radian = 6371.0088
            epsilon = kms / kms_per_radian
            db = DBSCAN(eps=epsilon, min_samples=min_no, algorithm='ball_tree', metric='haversine').fit(coords)
            core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
            #db.labels_ is a list containing the labels for all the points (118 in this case) as to which cluster they belong to
            #The labels are from 0 to the total number of clusters. The outliers are given a label of -1
            #np.zeros_like() gives a list of zeros with the same shape of the list mentioned (db.labels_ in this case)
            #The additional parameter dtype=bool converts each value to False since all of them are zeros
            core_samples_mask[db.core_sample_indices_] = True
            #db.core_sample_indices_ gives all the indices of db at which the values are not -1
            #core_samples_mask[db.core_sample_indices_] = True converts all values that were False at these indices to True
            #We now have core_samples_mask which is a list containing boolean values with True at points belonging to a cluster
            #and False at points that are outliers
            labels = db.labels_
            #We then save all the labels in a variable called labels
            # Number of clusters in labels, ignoring noise if present.
            n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
            #The number of clusters is stored in n_clusters_ 
            #Note that the -1 is not a cluster and hence is not counted in this

            print('Estimated number of clusters: %d' % n_clusters_)

            #forming a temporary list to store all labels and label -1 as well for outliers if present to plot outliers as well
            temp_labels=[]
            for i in range(n_clusters_):
                temp_labels.append(i)
            if -1 in labels:
                temp_labels.append(-1)

            clusters = pd.Series([coords[labels == n] for n in temp_labels])#range(n_clusters_)])


            main_list=[]
            for i in range(len(clusters)):
                Lat_list=[]
                Long_list=[]
                for j in range(len(clusters[i])):
                    Lat_list.append(Boroughs[borough_name][year][0][np.where(coords==[clusters[i][j]])[0][0]])
                    Long_list.append(Boroughs[borough_name][year][1][np.where(coords==[clusters[i][j]])[0][0]])
                    #np.where(coords==[clusters[i][j]]) provides us with a tuple containing two arrays like this
                    #(array([65, 65], dtype=int64), array([0, 1], dtype=int64))
                    #The first array contains the index of the list coords at which the jth radian coordinates of the ith cluster is.
                    #The next two [0][0] let us reach that index in FatalityYear[year][lat/long]
                main_list.append([Long_list,Lat_list])
                #Add these lists of lat and long in a list for each cluster and finally add them to a main_list


            from mpl_toolkits.basemap import Basemap

            
            if borough_name=='MANHATTAN':
                ll_lat=40.67369472204318 
                ll_long= -74.06385898590088
                ur_lat = 40.878391204189704 
                ur_long=-73.88282060623169
            elif borough_name=='BROOKLYN':
                ll_lat= 40.54215634446196
                ll_long= -74.07851457595825
                ur_lat = 40.745309
                ur_long= -73.813359
            elif borough_name=='QUEENS':
                ll_lat= 40.472330
                ll_long= -74.094001
                ur_lat = 40.828342
                ur_long= -73.638429
            elif borough_name=='BRONX':
                ll_lat= 40.779405
                ll_long= -73.967279
                ur_lat = 40.922497
                ur_long= -73.731760
            elif borough_name=='STATEN ISLAND':
                ll_lat= 40.465861
                ll_long= -74.320549
                ur_lat =  40.663817
                ur_long= -74.001259

            m= Basemap(projection='mill', llcrnrlat = ll_lat, llcrnrlon = ll_long,
                                  urcrnrlat=ur_lat,
                                  urcrnrlon = ur_long, epsg=2263, resolution = 'h')
            unique_labels = list(set(labels))
            colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
            m.arcgisimage(service='ESRI_StreetMap_World_2D', xpixels = 12000, verbose= True)
            m.drawcoastlines()
            m.drawcountries(linewidth=2)
            m.drawcounties()
            m.drawrivers()

            for i,col in zip(range(len(main_list)), colors):
                if unique_labels[i] == -1:
                    # Black used for noise.
                    col = 'k'
                xpt, ypt = m(main_list[i][0],main_list[i][1])
                m.plot(xpt, ypt, 'co', markersize =7,markerfacecolor=col)
            title = ("Cluster Visualization for " + borough_name + " for year " + str(year) )
            plt.title(title)
            plt.show()
        
        else:
            print("Please make correct choices")
            clustering_borough_fatality(kms, min_no)
    else:
        print("Please make correct choices")
        clustering_borough_fatality(kms, min_no)
    answer=input("Would you like to cluster anything else (Y/N):")
    if answer=='Y':
        clustering_borough_fatality(kms, min_no)
    else:
        return
```

```python
radius=input("What do you want the minimum radius to be in kms?: ")
min_fatalities=input("What should the minimum number of fatal accidents be in that radius to form a cluster?: ")
clustering_borough_fatality(float(radius), float(min_fatalities))
```
